```python
import os
from PIL import Image
# set path to folder containing images
folder_path = '/content/drive/Shareddrives/ESEARCH_Spring_2024/Data/wildfire/sentinel2_RGB'
# create list of image file names
file_names = os.listdir(folder_path)
# create list to store image shapes
shapes = []
# loop over image files and get their shapes
for file_name in file_names:
    # open image with PIL
    image_path = os.path.join(folder_path, file_name)
    with Image.open(image_path) as img:
        # get shape and add to list
        shape = img.size
        shapes.append(shape)
        print(f"Shape of {file_name}: {shape}")
# find largest image
largest_shape = max(shapes)
print(f"\nLargest image shape: {largest_shape}")
```

```python
import os
from PIL import Image
# set path to input and output folders
input_folder_path = '/content/drive/Shareddrives/ESEARCH_Spring_2024/Data/wildfire/sentinel2_RGB'
output_folder_path = '/content/drive/Shareddrives/ESEARCH_Spring_2024/Data/wildfire/sentinel2_128x128'
# create output folder if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)
# create list of image file names
file_names = os.listdir(input_folder_path)
# set new size for images
new_size = (128, 128)
# loop over image files and resize/crop
for file_name in file_names:
    # open image with PIL
    image_path = os.path.join(input_folder_path, file_name)
    with Image.open(image_path) as img:
        # check if the image is smaller than the new size
        if img.size[0] < new_size[0] or img.size[1] < new_size[1]:
            # create a new blank image with the new size and paste the original image onto it
            new_img = Image.new("RGBA", new_size, (0, 0, 0, 255))
            offset = ((new_size[0] - img.size[0]) // 2, (new_size[1] - img.size[1]) // 2)
            new_img.paste(img, offset)
        else:
            # create a new blank image with the new size and paste the resized image onto it
            new_img = Image.new("RGBA", new_size, (255, 255, 255, 0))
            offset = ((new_size[0] - img.size[0]) // 2, (new_size[1] - img.size[1]) // 2)
            new_img.paste(img, offset)
        # save the new image with the same name as the original, in the output folder
        new_file_path = os.path.join(output_folder_path, file_name)
        new_img.save(new_file_path)
# print message when done
print("All images cropped to", new_size, "and saved to", output_folder_path)
```

```python

import os
from PIL import Image
# Set path to input folder containing resized images
input_folder_path = '/content/drive/Shareddrives/ESEARCH_Spring_2024/Data/wildfire/sentinel2_128x128'
# Set path to output folder for corrected images
output_folder_path = '/content/drive/Shareddrives/ESEARCH_Spring_2024/Data/wildfire/Sentinel2_Corrected_Names'
# Create output folder if it doesn't exist
if not os.path.exists(output_folder_path):
    os.makedirs(output_folder_path)
# Create list of image file names
file_names = os.listdir(input_folder_path)
# Loop over image files and save with corrected names
for idx, file_name in enumerate(file_names, start=1):
    # Open image with PIL
    image_path = os.path.join(input_folder_path, file_name)
    with Image.open(image_path) as img:
        # Save the image with corrected name
        new_file_name = f"Sentinel2_image_{idx}.png"
        new_file_path = os.path.join(output_folder_path, new_file_name)
        img.save(new_file_path)
# Print message when done
print("All images saved with corrected names to", output_folder_path)
```
